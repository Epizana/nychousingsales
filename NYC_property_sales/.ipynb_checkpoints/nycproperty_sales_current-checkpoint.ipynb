```python
import numpy as np
import pandas as pd
import datetime
```

```python
manhattan = pd.read_csv('Manhattan.csv', header = 4)
bronx = pd.read_csv('Bronx.csv', header = 4)
brooklyn = pd.read_csv('Brooklyn.csv', header = 4)
queens = pd.read_csv('Queens.csv', header = 4)
staten_island = pd.read_csv('Staten Island.csv', header =4)
```

```python
manhattan.loc[:,'BOROUGH'] = 'Manhattan'
bronx.loc[:,'BOROUGH'] = 'Bronx'
brooklyn.loc[:,'BOROUGH'] = 'Brooklyn'
queens.loc[:,'BOROUGH'] = 'Queens'
staten_island.loc[:,'BOROUGH'] = 'Staten Island'
```

```python
bronx.rename({' SALE PRICE ':'SALE PRICE'}, axis = 1, inplace=True)
brooklyn.rename({' SALE PRICE ':'SALE PRICE'}, axis = 1, inplace=True)
queens.rename({' SALE PRICE ':'SALE PRICE'}, axis = 1, inplace=True)
staten_island.rename({' SALE PRICE ':'SALE PRICE'}, axis = 1, inplace=True)
```

**Cleaning 'Tax Class at Present' and 'Tax Class at Time of Sale'**

```python
#count how many nulls are in 'tax class at present'
np.sum(combined.loc[:,'TAX CLASS AT PRESENT'].isnull())
```

```python
#impute that for NaN Tax Class at Present, there is no change and it should be set equal to Tax Class at Sale.
combined.loc[:, 'TAX CLASS AT PRESENT'] = combined.loc[:,'TAX CLASS AT PRESENT'].fillna(combined.loc[:,'TAX CLASS AT TIME OF SALE'])
```

```python
#change 'tax class at present' to str first, after imputing to collect imputed values (ints) into the same object type.
combined.loc[:, 'TAX CLASS AT PRESENT'] = combined.loc[:,'TAX CLASS AT PRESENT'].astype(str)
#combined.loc[:, 'TAX CLASS AT TIME OF SALE'] = combined.loc[:,'TAX CLASS AT TIME OF SALE'].astype(str)
```

```python
#one tax class is object, the other is int. They should both be categorical as the tax class is a bracket segmentation.
combined.loc[:,'TAX CLASS AT TIME OF SALE'] = combined.loc[:,'TAX CLASS AT TIME OF SALE'].astype('category')
combined.loc[:,'TAX CLASS AT PRESENT'] = combined.loc[:,'TAX CLASS AT PRESENT'].astype('category')
```

```python
#now take a look at 'tax class at present' to see if there are any utility/office buildilngs (tax class 3/4 respectively).
tax_class = combined.loc[:,'TAX CLASS AT PRESENT'].unique().tolist()
tax_class
```

**Convert the various objects into more specific types. Otherwise counting nulls and zeros for filtering purposes won't work.**

```python
#Convert 'Apartment Number' to str and strip any extra whitespace
combined.loc[:,'APARTMENT NUMBER'] = combined.loc[:,'APARTMENT NUMBER'].astype(str).str.strip()
```

```python
#Convert 'Apartment Number' to str and strip any extra whitespace
combined.loc[:,'ADDRESS'] = combined.loc[:,'ADDRESS'].astype(str).str.strip()
```

```python
np.sum(combined.loc[:,'SALE PRICE'].isnull())
np.sum(combined.loc[:,'RESIDENTIAL UNITS'].isnull())
np.sum(combined.loc[:,'TOTAL UNITS'].isnull())
```

```python
#replace commas in SALE PRICE column with empty space and convert to string, then float.
combined['SALE PRICE'] = combined['SALE PRICE'].str.replace(',','').astype(float).astype(int)
```

```python
price = combined.loc[combined['SALE PRICE'].isnull()]
price
```

```python
combined['LAND SQUARE FEET'] = combined['LAND SQUARE FEET'].str.replace(',','').astype(float).astype(int)
combined['GROSS SQUARE FEET'] = combined['GROSS SQUARE FEET'].str.replace(',','').astype(float).astype(int)
```

```python
#convert sale date to datetime
combined.loc[:,'SALE DATE'] = pd.to_datetime(combined.loc[:,'SALE DATE'])
```

```python
combined['SALE MONTH'] = pd.DatetimeIndex(combined['SALE DATE']).month
```

```python
#residential units, commercial units, and total units are not consistent. Convert all to int.
combined.loc[:,'RESIDENTIAL UNITS'] = pd.to_numeric(combined.loc[:,'RESIDENTIAL UNITS'], errors = 'coerce')
combined.loc[:,'TOTAL UNITS'] = pd.to_numeric(combined.loc[:,'TOTAL UNITS'], errors = 'coerce')
```

```python
#residential units, commercial units, and total units are not consistent. Convert all to int.
combined.loc[:,'RESIDENTIAL UNITS'] = combined.loc[:,'RESIDENTIAL UNITS'].astype(int)
combined.loc[:,'TOTAL UNITS'] = combined.loc[:,'TOTAL UNITS'].astype(int)
```

### **Filter out non-residential transactions: office/utility/hotel buildings. Also filter out rental-only buildings: these are bought and sold by corporations and not individuals.**

```python
#filter out office buildings and rental buildings, sort by price descending. Looks like there are dupes.
combined = combined.loc[(combined['TAX CLASS AT PRESENT'] != '4') & (combined['BUILDING CLASS CATEGORY'].str.contains('RENTAL')==False)].sort_values('SALE PRICE', ascending = False)
```

**Filter out transfers (price is <= 10) and transactions under $100k**

```python
#filter price
combined = combined.loc[combined['SALE PRICE'] >= 100000,]
```

```python
#CREATE 'TYPE' COLUMN WHICH SHOWS THE BUILDING TYPE (CONDO VS CO-OP VS OTHER)
combined['TYPE'] = np.where(combined['BUILDING CLASS CATEGORY'].str.contains('CONDO COOPS'),\
                'Condop', np.where(combined['BUILDING CLASS CATEGORY'].str.contains('CONDO'),\
                'Condo',np.where(combined['BUILDING CLASS CATEGORY'].str.contains('COOP'),'Co-op','Other')))
```
