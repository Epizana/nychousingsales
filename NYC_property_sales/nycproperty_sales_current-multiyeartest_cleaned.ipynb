## **Address Processing**

```python
#take away extra spaces that may be in some rows under ADDRESS
#combined['ADDRESS'] = combined['ADDRESS'].str.strip()
```

```python
combined = combined.loc[combined["NEIGHBORHOOD"].str.contains('UNKNOWN')==False]
```

```python
combined['Base_Address'] = np.where(combined['ADDRESS'].str.contains(',NY,NY|.NY,NY|,NY'),combined['ADDRESS'].str.split(',').str[0],combined['ADDRESS'])
#test1['addresstest'] = np.where(test1['ADDRESS'].str.contains('.NY'),test1['ADDRESS'].str.split('.').str[0],test1['ADDRESS'])
```

```python
combined['Base_Address'] = combined['Base_Address'].str.strip().str.replace('NY,NY| NY,NY|NY, NY|, NY|NY, N|NY','', regex = True)
```

```python
combined['Base_Address'] = np.where(combined['Base_Address'].str.contains('.NY'),combined['Base_Address'].str.split('.').str[0],combined['Base_Address'])
```

```python
combined['Base_Address'] = np.where(combined['Base_Address'].str.contains('STREET') == False, combined['Base_Address'].str.strip().str.replace('ST.',' STREET', regex = False),combined['Base_Address'])
```

```python
combined['Base_Address'] = np.where(combined['Base_Address'].str.contains('STREET') == False, combined['Base_Address'].str.strip().str.replace(' ST',' STREET', regex = False),combined['Base_Address'])
```

```python
#Base_Address will be used to populate Building Address (excludes Apt. No) and Main_Address, which
#is either Building Address or Building Address plus Apt. No (which was extracted from either the Address
#or Apartment Number columns, whichever contained. If both contained, Apartment Number has the priority.
#this will be used for apartment sales filtering later on.

combined['Building_Address'] = combined['Base_Address'].str.split(',').str[0].str.strip()
combined['Apt_No'] = combined['Base_Address'].str.split(',').str[1]
combined['Apt_No'][combined['Apt_No'].isnull()] = combined['APARTMENT NUMBER'][combined['Apt_No'].isnull()]
combined['Apt_No'][(combined['Apt_No'].isnull()) | (combined['Apt_No'] =='nan')] = ''
combined['Main_Address'] = (combined['Building_Address'] + ', ' + combined['Apt_No']).str.rstrip(', ')

#back to null
combined['Apt_No'][combined['Apt_No']=='']=np.nan
```
